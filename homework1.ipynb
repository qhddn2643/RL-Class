(b) Similarly, implement a *UCB* policy class. Compared to the random policy class, the only difference would be in the arm selection strategy where you should use the UCB strategy.

```python
class UCBPolicy:
    def __init__(self, K, c) -> None:
        self.K = K
        self.c = c
        self.reset()

    def reset(self):
        #--- Your code here ---#
        self.N = np.zeros(self.K)
        self.Q = np.zeros(self.K)

    def choose_arm(self):
        #--- Your code here ---#
        # exploit
        for arm in range(self.K):
            if self.N[arm] == 0:
                return arm
        
        ucb_values = [0.0 for arm in range(self.K)]
        total_counts = sum(self.N)
        for arm in range(self.K):
            bonus = self.c * np.sqrt(np.log(total_counts) / np.float(self.N[arm]))
            ucb_values[arm] = self.Q[arm] + bonus
        return np.argmax(ucb_values)
    

    def update_estimate(self, k, r):
        #--- Your code here ---#
        self.N[k] = self.N[k] + 1
        self.Q[k] += np.sum(r - self.Q) / self.N[k]
        return self.Q[k]
    
```
