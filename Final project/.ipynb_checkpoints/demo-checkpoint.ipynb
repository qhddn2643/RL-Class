```python
import gym
import numpy as np
import matplotlib.pyplot as plt

import sys
import os
sys.path.append(os.path.abspath("/Users/qhddn/RL class/Final project"))
import final_project
```

```python
env = gym.make('Acrobot-v1')
```

```python
sum_rewards = []
num_step = []
policies = []
num_eps = 500
n_runs = 10
avg_rewards = np.zeros(num_eps)
avg_step = np.zeros(num_eps)

for i in range(n_runs):
    acrobot = final_project.Actor_Critic_Eligibility_Traces(env, n_episode=num_eps, lamda=0.1, alpha=0.5, gamma=0.9)
    policy, rewards, step = acrobot.play()
    policies.append(policy)
    sum_rewards.append(rewards/num_eps)
    num_step.append(step)

sum_rewards = np.array(sum_rewards)
num_step = np.array(num_step)

max_reward = np.max(sum_rewards)
min_step = np.min(num_step)

print("Maximum Reward:", max_reward)
print("Minimum Step:", min_step)

for i in range(len(sum_rewards)):
    avg_rewards[i] = np.average(sum_rewards[i])
    
for i in range(len(num_step)):
    avg_step[i] = np.average(num_step[i])

plt.plot(avg_rewards)
plt.xlabel("Episodes")
plt.ylabel("Average Rewards")
plt.title("Total reward on episode vs Number of episodes")
plt.show()    
    
plt.plot(avg_step)
plt.xlabel("Episodes")
plt.ylabel("Average Steps")
plt.title("Step-per-episode vs Number of episodes")
plt.show()
```
