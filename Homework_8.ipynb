## Part I (40 points)

Implement ACTOR-CRITIC with Eligibility Trace (p. 332).



```python
# Your code here (you are allowed to import from an external python file of your own impmenetation 
# instead of copying all the code here)
from collections import defaultdict
import itertools

def fourier_basis(S, n):
    if type(S) == "<class 'float'>": # if S is float
        k = 1
    else: # if S is 2D-array
        k = 2
    X = np.array([np.cos(i * np.pi * S) for i in range((n+1)**k)])
    return X

def policy(states):
    x = np.maximum(0, states)
    action_probs = np.exp(x) / np.sum(np.exp(x), axis=0)
    action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
    return action

def stateValue(S, n, w):
    V = w * fourier_basis(S, n)
    return V


def Actor_Critic_Eligibility_Traces(env, n_episode, lamda, alpha, gamma, tag):
    pi = defaultdict(lambda: np.zeros(env.action_space.n))
    V = defaultdict(float)
    num_step = np.zeros(n_episode)
    sum_reward = np.zeros(n_episode)
    w = np.zeros(len(fourier_basis(env.observation_space.shape[0], 3)))
    theta = np.zeros(len(fourier_basis(env.observation_space.shape[0], 3)))
    score = 0
    
    for episode in range(n_episode):
        
        if tag == "mc":
            S, info = env.reset()
        else:
            S = env.reset()
            S = S[0]
            
        z_theta = 0
        z_w = 0
        I = 1
        
        terminated = False
        
        for t in itertools.count():
            
            # Animate the last episode
            if episode == n_episode - 1:
                env.render()
                
            A = policy(pi[S])
            observation, R, terminated, truncated = env.step(A)
            
            if tag == "mc":
                next_S = observation[0], observation[1]
            else:
                next_S = observation[0], observation[1], observation[2], observation[3]
                
            score += R
            sum_reward[episode] = score
            num_step[episode] = t

            delta = R + (gamma * stateValue(V[next_S], 3, w)) - stateValue(V[S], 3, w)
            
            if terminated:
                V[next_S] = 0
                break
                
            z_w = (gamma * (lamda**w) * z_w) + fourier_basis(V[S], 3)
            z_theta = (gamma * (lamda**theta) * z_theta) + (I * np.log(pi[S][A]))
            w += (alpha**w) * delta * z_w

            theta += (alpha**theta) * delta * z_theta
            
            I = gamma * I
            S = next_S
                
    return sum_reward, num_step

```

## Part II (60 points)

Use the algorithm to learn both tasks. Tune the step-size parameter ($\alpha$), the Function Approximation order, the discount factor ($\gamma$), and the $\lambda$-value. **Note:** you can consider the problem to be undiscounted.

For each task:
 
1. Plot step-per-episode (in log scale) vs. number of episodes. This plot should be averaged over 50-100 runs. 
2. Plot total reward on episode vs. number of episodes. This plot should be averaged over 50-100 runs.
3. Show an animation of the task for the final episode.


```python
# Your code here
env = gym.make('MountainCar-v0')

sum_rewards = []
num_step = []
num_eps = 100
n_runs = 100
avg_rewards = np.zeros(num_eps)
avg_step = np.zeros(num_eps)

for i in range(n_runs):
    rewards, step = Actor_Critic_Eligibility_Traces(env, n_episode=num_eps, lamda=0.1, alpha=0.5, gamma=0.9, tag="mc")
    sum_rewards.append(rewards/num_eps)
    num_step.append((10**step))

sum_rewards = np.array(sum_rewards)
num_step = np.array(num_step)

for i in range(50, len(sum_rewards)):
    avg_rewards[i] = np.average(sum_rewards[i])
    
for i in range(50, len(num_step)):
    avg_step[i] = np.average(num_step[i])

    
plt.plot(avg_step)
plt.xlabel("Episodes")
plt.ylabel("Average Steps (in log scale)")
plt.title("Step-per-episode vs Number of episodes")
plt.show()


plt.plot(avg_rewards)
plt.xlabel("Episodes")
plt.ylabel("Average Rewards")
plt.title("Total reward on episode vs Number of episodes")
plt.show()
```
